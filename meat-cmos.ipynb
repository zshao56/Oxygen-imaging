```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import rotate
from matplotlib.colors import LinearSegmentedColormap
```

```python
# Load 2D array data from a text file
file_path = r"F:\0214\meat\meat-04_intensity_matrix.txt"
data = np.loadtxt(file_path)

value = data[]
```

```python
# Define cropping ranges
xmin, xmax = 0, 5000  # X direction cropping range
ymin, ymax = 0, 5000  # Y direction cropping range

# Crop the data
cropped_reflection = data[ymin:ymax, xmin:xmax]

# Mirror the cropped data horizontally
mirrored_reflection_horizontal = np.fliplr(cropped_reflection)

# Mirror the cropped data vertically
mirrored_reflection_vertical = np.flipud(cropped_reflection)

# Rotate the vertically mirrored data
rotated_mirrored_reflection_vertical = rotate(mirrored_reflection_vertical, 0, reshape=False)

```

```python

# Define custom colormap from white to blue
R = 30
G = 70
B = 110
R = R/256
G = G/256
B = B/256
colors = [(R, G, B), (1, 1, 1)]  
cmap = LinearSegmentedColormap.from_list('custom_cmap', colors)

# Plot the rotated vertically mirrored data as a heatmap
plt.imshow(rotated_mirrored_reflection_vertical, cmap=cmap, origin='lower', vmin=0, vmax=270)

# Add colorbar
plt.colorbar(label='intensity')

# Add title and labels
plt.xlabel('X')
plt.ylabel('Y')

plt.ylim(0, 1200)
plt.xlim(250, 1450)
# Save the plot as a PDF file
output_file_path = r'F:\OneDrive - UW-Madison\python\figure&summary\output pdf\meat04.pdf'
plt.savefig(output_file_path, format='pdf', bbox_inches='tight')
plt.show()
```
